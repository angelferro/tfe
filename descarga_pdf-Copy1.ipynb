```python
page=requests.get("http://www.congreso.es/portal/page/portal/Congreso/Congreso/Publicaciones?_piref73_2342619_73_1340041_1340041.next_page=/wc/servidorCGI&CMD=VERLST&CONF=BRSPUB.cnf&BASE=PU12&FMT=PUWTXLTS.fmt&DOCS=1-25&DOCORDER=FIFO&OPDEF=Y&QUERY=%28D%29.PUBL.")
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
vector = []
```

```python
for link in soup.findAll('a',href=re.compile(r"\.CODI.#1$")):
    vector.append("http://www.congreso.es" + str(link.get('href')))
```

```python
len(vector)
```

```python
vector = vector [25:50]
```

```python
textos=[]
for l in range(len(vector)):
    p=requests.get(vector[l])
    s = BeautifulSoup(p.content,'html.parser')
    textos.append(s)
```
